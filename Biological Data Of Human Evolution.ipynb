### Investigating the possible linearity of evolution over time.

```python
ascending_time = data['Time'].sort_values()

plt.bar(ascending_time, Cranial_Capacity, color='green')

plt.xlabel('Time')
plt.xticks(rotation=70)
plt.ylabel('Cranial Capacity')
plt.title('Cranial Capacity over Time')
```

```python
plt.scatter(ascending_time, Cranial_Capacity, color='green',s=20)

plt.xlabel('Time')
plt.xticks(rotation=70)
plt.ylabel('Cranial Capacity')
plt.title('Cranial Capacity over Time')
```

```python
plt.scatter(ascending_time, height, color='orange')

plt.xlabel('Time')
plt.xticks(rotation=70)
plt.ylabel('Height')
plt.title('Height over Time')
```

```python
plt.bar(ascending_time, Anatomy, color='lightgreen')

plt.xlabel('Time')
plt.xticks(rotation=70)
plt.ylabel('Anatomy')
plt.title('Anatomy over Time')
```

```python
plt.bar(ascending_time, biped, color='green')

plt.xlabel('Time')
plt.xticks(rotation=70)
plt.ylabel('Bipedalism')
plt.title('Bipedalism over Time')
```

The only pattern we can see based on the results is the height has increased over time, but if we check other columns related in hte dataset like the anatomy or the bipedalism doesn´t have a relationship.

is related to a very modern jaw shape, the moderm bipedalism, the modern foramen magnum position and with the walking foot type but there isn't
